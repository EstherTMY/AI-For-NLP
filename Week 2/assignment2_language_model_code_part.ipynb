```python
import os
from hanziconv import HanziConv
```

```python
database_path = 'text/AA' # computer performance limited, only process AA folder's data
```

```python
files= os.listdir(database_path)
```

```python
content = []
for file in files:
    if not os.path.isdir(file):
        f = open(database_path+"/"+file)
        iter_f = iter(f)
        str = ""
        for line in iter_f:
            str = str + line
        content.append(str)
```

```python
content[:1]
```

```python
import re
```

```python
def token(string):
    return ' '.join(re.findall('[\w|\d]+', string))
```

```python
all_articles = [token(a) for a in content]
```

```python
all_articles = [HanziConv.toSimplified(a) for a in all_articles]
```

```python
all_articles[:1]
```

```python
text = ''

for a in all_articles:
    text += a
print('length of text: {}'.format(len(text)))
```

```python
TEXT = text
```

```python
import jieba
```

```python
def cut(string): return list(jieba.cut(string))
```

```python
ALL_TOKENS = cut(TEXT)
```

```python
valid_tokens = [t for t in ALL_TOKENS if t.strip()]
```

```python
len(valid_tokens)
```

```python
valid_tokens[:20]
```

```python
from collections import Counter
```

```python
words_count = Counter(valid_tokens)
```

```python
words_count.most_common(10)
```

```python
frequences_all = [f for w, f in words_count.most_common()]
```

```python
frequences_sum = sum(frequences_all)
```

```python
1 / frequences_sum
```

```python
def get_prob(word): 
    esp = 1 / frequences_sum
    if word in words_count: 
        return words_count[word] / frequences_sum
    else:
        return esp
```

```python
get_prob('的')
```

```python
def product(numbers):
    return reduce(lambda n1, n2: n1 * n2, numbers)
```

**1 - gram language model**

```python
def language_model_one_gram(string):
    words = cut(string)
    return product([get_prob(w) for w in words])
```

```python
from functools import reduce
```

```python
language_model_one_gram('广交会下个月举办')
```

**2 - gram language model**

```python
all_2_grams_words = [''.join(valid_tokens[i:i+2]) for i in range(len(valid_tokens[:-2]))]
```

```python
_2_gram_sum = len(all_2_grams_words)
_2_gram_counter = Counter(all_2_grams_words)
```

```python
def get_combination_prob(w1, w2):
    if w1 + w2 in _2_gram_counter: return _2_gram_counter[w1+w2] / _2_gram_sum
    else:
        return 1 / _2_gram_sum
```

```python
get_combination_prob('去', '北京')
```

```python
def get_prob_2_gram(w1, w2):
    return get_combination_prob(w1, w2) / get_prob(w1)
```

```python
get_prob_2_gram('去', '沈阳')
```

```python
get_prob_2_gram('去', '北京')
```

```python
def langauge_model_of_2_gram(sentence):
    sentence_probability = 1
    
    words = cut(sentence)
    
    for i, word in enumerate(words):
        if i == 0: 
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous, word)
        sentence_probability *= prob
    
    return sentence_probability
```

```python
langauge_model_of_2_gram('小明今天抽奖抽到一台苹果手机')
```

```python
langauge_model_of_2_gram('小明今天抽奖抽到一台波音飞机')
```
